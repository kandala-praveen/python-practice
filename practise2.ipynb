```python
# largest element in list
n=int(input("enter the length of list"))
a=[]
for i in range(n):
    x=int(input(f"enter the {i} element:"))
    a.append(x)
print(a)
for i in range(n):
    for j in range(n-1):
        if a[j]>a[j+1]:
            a[j],a[j+1]=a[j+1],a[j]
print("sorted list:",a)
print("largest element in list:",a[-1])
```

```python
#duplicate removing
n=int(input("enter the length of list:"))
a=[]
for i in range(n):
    x=int(input(f"enter the {i} element:"))
    a.append(x)
print(a)
for i in range(n):
    for j in range(n-1):
        if a[j]>a[j+1]:
            a[j],a[j+1]=a[j+1],a[j]
print("sorted list:",a)
i=0
while i<(len(a)-1):
    if a[i]==a[i+1]:
        del a[i]
    else:
        i+=1
print("removed duplicates:",a) 
```

```python
#reverse the list
n=int(input("enter the length of list:"))
a=[]
for i in range(n):
    x=int(input(f"enter the {i} element:"))
    a.append(x)
print(a)
i=0
j=len(a)-1
while i<j:
    a[i],a[j]=a[j],a[i]
    i+=1
    j-=1
print("reserved list:",a)
```

```python
#merge the lists
n1=int(input("enter the length of list1:"))
l1=[]
for i in range(n1):
    x=input(f"enter the {i} element:")
    l1.append(x)
n2=int(input("enter the length of list2:"))
l2=[]
for j in range(n2):
    y=input(f"enter the {j} element:")
    l2.append(y)
print("list1",l1)
print("list2:",l2)
l1.extend(l2)
print(l1)
```

```python
#even and odd elements seperate
n=int(input("enter the length of list:"))
li=[]
for i in range(n):
    x=int(input("enter the element:"))
    li.append(x)
print(li)
l1=[]
l2=[]
for j in range(len(li)):
    if li[j]%2==0:
        x=li[j]
        l1.append(x)
    else:
        y=li[j]
        l2.append(y)
print("even numbers",l1)
print("odd numbers",l2)
```
