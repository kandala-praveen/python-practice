# file handling

# creating a file

```python
# open("filename","permision")
s="""this is my file handling program.
it is a concept in python programming language.
the real examples of file handling are attaching the in whatsapp,teams
file handling has permision like "r","a","w" which are used to read,append,write the files"""
file=open("demo1.txt","w")
file.write(s)
file.close()
print("file created")
```

# read from file

```python
#open("filename","r")
file=open("demo1.txt","r")
print(file.read())
```

# write a list into file

```python
#write a list into file
li=["python","java","angular","c"]
file=open("demo2.txt","w")
file.writelines(li)
file.close()
print("file created")
```

```python
# read li from file
file=open("demo2.txt","r")
filecontent=file.readlines()
print(filecontent)
```

```python
# appending the list
s="python file handling"
file=open("demo2.txt","a")
file.write(s)
file.close()
```

```python
#read N lines form file
n=int(input("enter the no of lines:"))
f=open("demo1.txt","r")
for i in range(n):
    line=f.readline()
    print(line)
f.close()
```
