# conditional statements

```python
# simple if statement (checks single statement)
x=65
if x==65:
    print("yes! x is 65.")
```

```python
#if else statement (checks two conditions)
x=12
if (x%2==0):
    print("yes")
else:
    print("no")
```

```python
# id else ladder
x=12
y=23
z=2
if x>y and x>z:
    print("x is bigger")
elif y>z and y>x:
    print("y is bigger")
else:
    print("z is bigger")
```

```python
#nested if condition
age=int(input("enter a age of a person:"))
weight=int(input("enter the weight of a person:"))
if age>=18:
    if weight>=50:
        print("can donate blood")
    else:
        print("weight is lesser,so cant donate blood")
else:
    print("can't donate blood")
```

# control statements
# -->simple for loop
# -->while loop
# -->nested loops

```python
# for loop
for i in range(1,10,1):# range(start,end-1,step size)
    print(i,end=" ") # end function is used for horizontal result
```

```python
for i in range(1,6):
    for j in range(i):
        print("*",end="")
    print()
```

```python
for i in range(5,0,-1):
    for j in range(i):
        print("*",end="")
    print()
```

# looping statements

```python
for letter in "python":
    if letter =="o":
        #break is used to break the execution
        break
    print(letter,end=" ")
print("bye")
```

```python
for letter in "python":
    if letter =="o":
        #continue is used to skip the particular execution
        continue
    print(letter,end=" ")
print("bye")
```

```python
for letter in "python":
    if letter =="o":
        pass
    print(letter,end=" ")
print("bye")
```
