```python
#Create BankAccount class with deposit() & withdraw()
class bank:
    def __init__(self,holdername,deposit,withdrawal):
        self.holdername=holdername
        self.balance=10000
        self.deposit=deposit
        self.withdrawal=withdrawal
    def depo(self):
        self.balance=self.balance+self.deposit
        print("total balance:",self.balance)
    def withd(self):
        if self.balance<self.withdrawal:
            print("balance is less")
        else:
            self.balance=self.balance-self.withdrawal
            print("total balance:",self.balance)
    def display(self):
        print("holdername:",self.holdername)
        print("balance:",self.balance)
holdername=input("enter the holdername:")
deposit=int(input("enter the deposting amount:"))
withdrawal=int(input("enter the withdraw amount:"))
x=bank(holdername,deposit,withdrawal)
x.display()
x.depo()
x.withd()
```

```python
# Shopping cart
class shopping:
    def __init__(self):
        self.total=0
    def calculate(self,items,cost):
        self.total=self.total+cost
    def display(self):
        print("total cost:",self.total)
x=shopping()
n=int(input("enter the no of items:"))
for i in range(n+1):
    items=input("enter the item:")
    cost=int(input("enter the cost:"))
    x.calculate(items,cost)
x.display()
```

```python
# employees salary system
class employee:
    def __init__(self,name,msalary):
        self.name=name
        self.msalary=msalary
        self.ysalary=0
        self.bon=0
        print("name:",self.name)
        print("month:",self.msalary)
    def yearly_salary(self):
        self.ysalary=self.msalary*12
        print("yearly:",self.ysalary)
    def bonus(self):
        if self.msalary>=50000:
            self.bon=(self.msalary)//10
            print("bonus:",self.bon)
        else:
            self.bon=(5*self.msalary)//100
            print("bonus:",self.bon)
name=input("enter the name:")
msalary=int(input("enter the monthly salary:"))
x=employee(name,msalary)
x.yearly_salary()
x.bonus()
```

# inheritance

```python
# parent and child
class person:
    def __init__(self,name,age):
        self.name=name
        self.age=age
        print("name:",self.name)
        print("age:",self.age)
class student(person):
    def __init__(self,name,age,marks):
        super().__init__(name,age)
        self.marks=marks
        print("marks",self.marks)
name=input("enter the name:")
age=int(input("enter the age:"))
marks=int(input("enter the marks:"))
x=student(name,age,marks)
```

```python
# employee-manager
class employee:
    def __init__(self,name,salary):
        self.name=name
        self.salary=salary
    def display_emp(self):
        print("name:",self.name)
        print("salary:",self.salary)
class manager(employee):
    def __init__(self,name,salary):
        super().__init__(name,salary)
        self.bonus=(self.salary//10)
        print("bonus:",self.bonus)
        self.total=self.salary+self.bonus
    def display_total(self):
        print("total:",self.total)
name=input("enter the name:")
salary=int(input("enter the salary:"))
x=manager(name,salary)
x.display_emp()
x.display_total()
```

```python
# mutlilevel
class animal:
    def eat(self):
        print("animal is eating")
class dog(animal):
    def bark(self):
        print("dog is barking")
class puppy(dog):
    def weep(self):
        print("puppy is weeping")
x=puppy()
x.eat()
x.bark()
x.weep()
```

```python
# device-mobile-smart phone
class device:
    def device_info(self):
        print("electronic device")
class mobile(device):
    def calling(self):
        print("Calling feature available")
class smart(mobile):
    def internet(self):
        print("Internet feature available")
s=smart()
s.device_info()
s.calling()
s.internet()
```

```python
#hierarchical inheritance
class employee:
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def display(self):
        print("name:",self.name)
        print("age:",self.age)
class developer(employee):
    def coding(self):
        print("Developer writes code")
class tester(employee):
    def testing(self):
        print("Tester tests software")
name=input("enter the name:")
age=int(input("enter the age:"))
x=developer(name,age)
y=tester(name,age)
y.testing()
x.coding()
```
