```python
# palindrome of lists
n=int(input("enter the length of list:"))
l=[]
for i in range(n):
    x=int(input(f"enter the {i} element:"))
    l.append(x)
z=l.copy()
print("current list:",z)
i=0
j=len(l)-1
while i<j:
    l[i],l[j]=l[j],l[i]
    i+=1
    j-=1
print("reversed list",l)
if z==l:
    print("palindrome")
else:
    print("non palindrome")
```

```python
#liner search
n=int(input("enter the length of list:"))
l=[]
for i in range(n):
    x=int(input(f"enter the {i} element:"))
    l.append(x)
print("current list:",l)
s=int(input("enter the search element:"))
for i in range(n):
    if l[i]==s:
        print(f"element found at {i}th index")
```

```python
#frequency of each element
n=int(input("enter the length of list:"))
l=[]
for i in range(n):
    x=int(input(f"enter the {i} element:"))
    l.append(x)
print("current list:",l)
for i in range(len(l)-3):
    count=0
    s=int(input("enter the search element:"))
    for j in range(n):
        if l[j]==s:
            count+=1
    print(f"the {s} element repeats : {count} times")
```

```python
#sum of elements in list
n=int(input("enter the length of list:"))
l=[]
for i in range(n):
    x=int(input(f"enter the {i} element:"))
    l.append(x)
print("current list:",l)
count=0
for i in range(len(l)):
    count+=l[i]
print("total sum of list:",count)
```

# tuples

```python
#converting tuple into list
n=int(input("enter the length of tuple:"))# cant change tuple during run time so by list only we can access one by one elements
l=[]
for i in range(n):
    x=int(input(f"enter the {i} element"))
    l.append(x)
print("current list:",l)
t1=tuple(l)
print("curren tuple",t1)
l1=list(t1)
print("after convertion",l1)
```

```python
# find max and min values
n=int(input("ënter the length of list:"))
l=[]
for i in range(n):
    x=int(input(f"enter the {i} element:"))
    l.append(x)
print("current list:",l)
for i in range(n):
    for j in range(n-1):
        if l[j]>l[j+1]:
            l[j],l[j+1]=l[j+1],l[j]
t1=tuple(l)
print("sorted tuple:",t1)
print("maximum element:",t1[-1])
print("minimum element:",t1[0])
```

```python
#reserve the tuple
t=tuple(map(int,input("enter the elements:").split()))
print("current tuple:",t)
l=list(t)
i=0
j=len(t)-1
while i<j:
    l[i],l[j]=l[j],l[i]
    i+=1
    j-=1
x=tuple(l)
print("reversed tuple:",x)
```

```python
#merging of two tuples
t1=tuple(map(int,input("enter the elements for t1:").split()))
t2=tuple(map(int,input("enter the elements for t2:").split()))
t3=()
t3=t1+(t2)
print("merged tuple:",t3)
```

```python
#count occurances in tuples
t=tuple(map(int,input("enter the elements:").split()))
l=list(t)
x=int(input("ënter the occured element:"))
count=0
for i in range(len(l)):
    if l[i]==x:
        count+=1
print(count)
```

```python
#duplicates removing
t=tuple(map(int,input("enter the elements:").split()))
l=list(t)
for i in range(len(l)):
    for j in range(len(l)-1):
        if l[j]>l[j+1]:
            l[j],l[j+1]=l[j+1],l[j]
print("current list:",l)
i=0
while i<len(l)-1:
    if l[i]==l[i+1]:
        del l[i]
    else:
        i+=1
x=tuple(l)
print("sorted tuple:",x)
```

# dictionaries

```python
#merging of two dictionaries
n=int(input("enter the length of dict:"))
d1={}
for i in range(n):
    key=input(f"enter the key {i}:")
    value=input(f"enter the value {i}:")
    d1[key]=value
print("current dict:",d1)
n=int(input("enter the length of dict:"))
d2={}
for i in range(n):
    key=input(f"enter the key {i}:")
    value=input(f"enter the value {i}:")
    d2[key]=value
print("current dict:",d2)
d1.update(d2)
print("merged dict:",d1)
```

```python
#search
n=int(input("enter the length of dict:"))
d={}
for i in range(n):
    key=input(f"enter the {i} key:")
    value=input(f"enter the {i}value:")
    d[key]=value
print("current dict:",d)
x=input("enter the key:")
if x in d:
    print("key found")
else:
    print("not found-")
```

```python
#reverse a dictionary
n=int(input("enter the length of dict:"))
d={}
for i in range(n):
    key=input(f"enter the {i} key:")
    value=input(f"enter the {i}value:")
    d[key]=value
print("current dict:",d)
l=list(d.items())
print(l)
i=0
j=len(l)-1
while i<j:
    l[i],l[j]=l[j],l[i]
    i+=1
    j-=1
print("reversed dict:",l)
```
