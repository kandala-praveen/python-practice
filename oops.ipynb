# creating a constructor

```python
class myclass():
    def __init__(self):
        print("this is constructor")
m=myclass()
```

```python
#changing value of object
class myclass():
    def __init__(self,name,age):
        print("this is a constructor")
        self.name=name
        self.age=age
    def myfunction(self): # creating a function
        print("my name is",self.name)
m1=myclass("praveen",21)
print(m1.name)
print(m1.age)
m1.myfunction()
#changing age in object
m1.age=20
print(m1.age)
#deleting of age
del m1.age
print(m1.age)
```

# inheritance


```python
#single inheritance
class parentclass():# parent class
    def myfunction1(self):# property of parent class
        print("parent class is called")
class childclass(parentclass):# child class
    def myfunction2(self):# property of child class
        print("child class is called")
m1=childclass()
m1.myfunction1()
```

```python
# multilevel inheritance has 3 levels of class
class parentclass():# parent class
    def myfunction1(self):# property of parent class
        print("parent class is called")
class subparentclass(parentclass):#sub parent class
    def myfunction2(self):# property of sub parent class
        print("sub parent class is called")
class childclass(subparentclass):#child class
    def myfunction3(self):# property of child class
        print("child class is called")
m1=childclass()
m1.myfunction1()
m1.myfunction2()
m1.myfunction3()
```

```python
# multiple inheritance
class parent1():
    def myfunction1(self):
        print("function1 is called")
class parent2():
    def myfunction2(self):
        print("function2 is called")
class c1(parent1,parent2):
    def myfunction3(self):
        print("function3 is called")
m1=c1()
m1.myfunction1()
m1.myfunction2()
m1.myfunction3()
```

```python
#herarchical inheritance
class parent():
    def myfunction(self):
        print("function1 is called")
class child2(parent):
    def myfunction2(self):
        print("function2 is called")
class child1(parent):
    def myfunction1(self):
        print("function3 is called")
m1=child1()
m2=child2()
m1.myfunction()
m1.myfunction1()
m2.myfunction()
m2.myfunction2()
```

```python
#hybrid inheritance
class parent:
    def myfunction(self):
        print("function is called")
class child1(parent):
    def myfunction1(self):
        print("function1 is called")
class child2(parent):
    def myfunction2(self):
        print("function2 is called")
class child3(child1,child2):
    def myfunction3(self):
        print("function3 is called")
m1=child3()
m1.myfunction()
m1.myfunction1()
m1.myfunction2()
m1.myfunction3()
```

```python
# polymorphism
# method overloading - same class,same function & different arguments
class myclass():
    def function(self):
        print("function 1 is called")
    def function(self,a):
        print("function 2 is called")
    def function(self,a,b):
        print("function 3 is called")
m=myclass()
m.function(10,20)
```

```python
# method overriding - same function,same arguments & different class
class myclass1():
    def function(self):
        print("function 1 is called")
class myclass2(myclass1):
    def function(self):
        print("function 2 is called")
        super().function()
class myclass3(myclass2):
    def function(self):
        print("function 3 is called")
        super().function()
m1=myclass3()
m1.function()
```
