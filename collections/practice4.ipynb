```python
# remove a key from dict
n=int(input("enter the length of dict:"))
d={}
for i in range(n):
    key=input(f"enter the {i} key:")
    value=input(f"enter the {i}value:")
    d[key]=value
print("current dict:",d)
x=input("enter the key:")
if x in d:
    d.pop(x)
print("updated dict:",d)
```

```python
# remove a key from dict
n=int(input("enter the length of dict:"))
d={}
for i in range(n):
    key=input(f"enter the {i} key:")
    value=int(input(f"enter the {i}value:"))
    d[key]=value
print("current dict:",d)
total=sum(d.values())
print("sum of values:",total)
```

```python
#sorting by values
n=int(input("enter the length of dict:"))
d={}
for i in range(n):
    key=input(f"enter the {i} key:")
    value=int(input(f"enter the {i}value:"))
    d[key]=value
print("current dict:",d)
x=sorted(d.values())
print("sorted values:",x)
```

# strings

```python
#print even & odd index characters
string=input("enter the string:")
for i in range(len(string)):
    if i%2==0 :
        print("even index characters:",string[i],end=" ")
    else:
        print("odd index characters:",string[i])
```

```python
#largest word in a sentence
n=input("enter the sentence:")
words=n.split()
largest=words[0]
print(words)
for i in words:
    if len(i)>len(largest):
        largest=i
print("largest word:",largest)      
```

```python
#smallest word in a sentence
n=input("enter the sentence:")
words=n.split()
print(words)
smallest=words[0]
for i in words:
    if len(i)<len(smallest):
        smallest=i
print("smallest word:",smallest)
```

```python
#replace spaces with -
n=input("enter the sentence:")
for i in n:
    if i==" ":
        n=n.replace(" ","-")
print(n)
```

```python
#maximum occuring character
n=input("enter the sentence:")
d={}
for i in n:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
max_occur=max(d,key=d.get)
print(d)
print(max_occur)
```

```python
#minimum occuring character
n=input("enter the sentence:")
d={}
for i in n:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
min_occur=min(d,key=d.get)
print(d)
print("minimum occuring character:",min_occur)
```

```python
#each character frequency
n=input("enter the sentence:")
d={}
for i in n:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
print("each character frequency:",d)
```

```python
#each word frequency
n=input("enter the sentence:")
words=n.split()
d={}
for i in words:
    if i in d:
        d[i]+=1
    else:
        d[i]=1
print("each word frequency:",d)
```

```python
#anagram
n1=input("enter the string:")
n2=input("enter the string:")
d1={}
d2={}
for i in n1:
    if i in d1:
        d1[i]+=1
    else:
        d1[i]=1
for j in n2:
    if j in d2:
        d2[j]+=1
    else:
        d2[j]=1
if d1==d2:
    print("anagram")
else:
    print("not anagram")
```
