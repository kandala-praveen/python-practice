# string slicing

```python
# access the index 5
a="hi, how are you?"
print("index 5:",a[5])
print("index 4 to last:",a[4:])
print("index first to 2",a[:2])
print("last index:",a[-1])
#updating the string
print("updated string:",a[:4]+"everyone")
```

# inbulit string methods

```python
# captialize: this method is used for convert the 1st letter of string into capital
a="python programming"
x=a.capitalize()
print(x)

#casefold : this method is used for convert all capitals into small letters
b="PRAVEEN Kandala"
y=b.casefold()
print(y)

#count: this method is used for count the char or strings repeated in string
z=y.count('a')
print(z)

#center: this method is used to center the string
s=y.center(50)
print(s)
```

# lists

```python
# creatung lists
li=[1,3,4,5,6,7,8,9]
print("created list:",li)
li=['pineapple','grapes','orange','mango']
print("created list:",li)
#list constructor
li=list((1,3,4,5,6,77,89,7))
print("list using list constructor:",li)
```

```python
# list allows duliplicate values
li=[12,344,667,2223,4443,12,4443,344,667]
print("duplicate valued list:",li)
```

```python
#list slicing 
li=[1,3,4,5,7,8,9,10]
#access the index of 3
print("index 3:",li[3])

#access index 1 to 4
print("index 1 to 4",li[1:4])

#access index till 5
print("index till 5",li[:6])

#access index from 4
print("index from 4",li[4:])

#access last index 
print("last index",li[-1])
```

```python
#extending list

l1=[12,23,45,56,75]
l2=["mango","grapes","orange","apple","pineapple","watermelon"]
l1.extend(l2)
print(l1)
```

```python
#removing items from list
# remove last element
l2=["mango","grapes","orange","apple","pineapple","watermelon"]
l2.pop()
print(l2)

#remove element from list by remove()
l2.remove("mango")
print(l2)

#del()
del l2[2]
print(l2)

#removing the all items from list
l2.clear()
print(l2)
```

```python
#updating list
l2=["mango","grapes","orange","apple","pineapple","watermelon"]
l2[1:3]=["guava","kiwi"]
print(l2)

#inserting into list
l2.insert(0,"kiwi")
print(l2)
```

# tuples

```python
# creating tuples
t1=("mango","orange","grapes",1,2,3,45,6,889,8,8.7,6.5,5,7)
print("created tuple:",t1)
#length of the tuple using len()
print("size:",len(t1))
```

```python
#indexing/slicing
#accesing index 1 value
t1=("mango","orange","grapes",1,2,3,45,6,889,8,8.7,6.5,5,7)
print("index 1 value of t1:",t1[1])

#accesing last index value
print("last index value:",t1[-1])

#accesping values between 0 and 6
print("index between 0 and 6 :",t1[0:6])

#accessing till 4
print("index till 4:",t1[:4])

#access from 6
print("index from 6:",t1[6:])
```

```python
# tuple constructor
t1=tuple((12,21,34,67,88,54))
print(t1)
```

```python
#converting into list and add a value
t1=(12,21,34,67,88,54)
print("current tuple:",t1)

#converting into list
x=list(t1)
print("after convert into list:",x)
print(type(x))

#add a value to list
x[2]=30
print("updated list:",x)

#converting list into tuple
y=tuple(x)
print("updated tuple:",y)
print(type(y))
```

```python
#loop with tuple
t1=(12,21,34,67,88,54)
for i in t1:
    print(i)
```

```python
#while loop with tuple
t1=(12,21,34,67,88,54)
i=0
while i<len(t1):
    print(t1[i])
    i+=1
else:
    print("loop finished")
```

```python
#comparing the tuples 
t1=(1,2,3,4)
t2=(1,2,4,5)
result= t1<t2
print(result)

res=t1>t2
print(res)
```

# dictionary

```python
#creating a dictionary
d1={1:"apple",2:"orange",3:"guava","hello":"namesthe"}
print("current dictionary:",d1)
print(type(d1))
```

```python
# list into data type
d1={1:"namaskaram",2:"namaskara",3:["hello","namasthey","hello"]}
print("current dict:",d1)

#tuple into dictionary
d1={1:"namaskaram",2:"namaskara",3:("hello","namasthey","hello")}
print("current dict:",d1)
```

```python
#accessing the values from dictionary
d1={1:"apple",2:"orange",3:"guava","hello":"namesthe"}
x=d1["hello"]
print("value of hello key:",x)

#access by get()
x=d1.get(1)
print("key value of 1:",x)

#access all values from dictionary
x=d1.values()
print("values of dictionary:",x)

#access of all keys from dictionary
y=d1.keys()
print("key of dict d1:",y)

#access by using items()
z=d1.items()
print(z)
```

```python
# removing values from dictionary
d1={1:"apple",2:"orange",3:"guava","hello":"namesthe"}
d1.pop(1)
print("remove key 1:",d1)

#removing using the del()
del d1[3]
print("updated dict:",d1)

#removing last item using the popitem()
d1.popitem()
print("updated dict:",d1)

#remving all values from dictionary
d1.clear()
print("updated dict:",d1)
```
