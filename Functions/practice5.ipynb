# functions

```python
# character frequency using dict
string=input("enter the string:")
d={}
def my(d1,st):
    for i in st:
        if i in d1:
            d1[i]+=1
        else:
            d1[i]=1
    print(d1)
my(d,string)
```

```python
# words frequency in string
string=input("enter the string:")
x=string.split()
d={}
def my(d1,x1):
    for i in x1:
        if i in d1:
            d1[i]+=1
        else:
            d1[i]=1
    print(d1)
my(d,x)
```

```python
#anagram check
x=input("ënter the string:")
y=input("enter the string:")
d1={}
d2={}
def my(x1,y1,d3,d4):
    for i in x1:
        if i in d3:
            d3[i]+=1
        else:
            d3[i]=1
    for j in y1:
        if j in d4:
            d4[j]+=1
        else:
            d4[j]=1
    if d3==d4:
        print("anagram")
    else:
        print("not anagram")
my(x,y,d1,d2)
```

```python
# power of a number
e1,p1=map(int,input("enter the exop and power:").split())
def my(e,p):
    r=1
    if e==1:
        return 1
    else:
        for i in range(p):
            r=r*e
        print(r)
print(my(e1,p1))
```

```python
#sum of n natural using recursion
x=int(input("enter a number:"))
def my(p):
    if p==1:
        return 1
    else:
        return p+my(p-1)
print(my(x))
```

```python
# Print numbers 1 to n using recursion
n = int(input("Enter a number: "))
def my(x):
    if x == 0:
        return
    my(x - 1)
    print(x)
my(n)
```

```python
#Fibonacci using functions
n=int(input("enter no of numbers:"))
def my(x):
    a=0
    b=1
    for i in range(n):
        print(a,end=" ")
        c=a+b
        a=b
        b=c
my(n)
```
