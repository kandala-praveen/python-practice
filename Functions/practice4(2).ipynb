```python
#even and odd 
def myfunction(n):
    if n%2==0:
        print(f"{n} is even number")
    else:
        print(f"{n} is odd number")
x=int(input("enter the number:"))
myfunction(x)
```

```python
#prime or not
def myfunction(n):
    count=0
    for i in range(1,n+1):
        if n%i==0:
            count+=1
    if count==2:
        print(f"{n} is prime number")
    else:
        print(f"{n} is non prime")
x=int(input("enter a number:"))
myfunction(x)
```

```python
#reserve a number
def my(n):
    res=0
    while n>0:
        temp=n%10
        res=(res*10)+temp
        n=n//10
    print(f"reverse of {n} is: ",res)
x=int(input("enter a number: "))
my(x)
```

```python
#square number
def my(n):
    return n*n
x=int(input("enter a number:"))
my(x)
```

```python
#sum of digits
def my(y):
    res=0
    while y>0:
        temp=y%10
        res=res+temp
        y=y//10
    print(f"sum of digits of {y}: ",res)
x=int(input("enter a number: "))
my(x)
```

```python
# palindrome
def my(n):
    y=n
    res=0
    while n>0:
        temp=n%10
        res=(res*10)+temp
        n=n//10
    if y==res:
        print("palindrome")
    else:
        print("non palindrome")
x=int(input("enter a number:"))
my(x)
```

```python
#primes in a range
def my(a,b):
    for i in range(a,b+1):
        count=0
        for j in range(1,i+1):
            if i%j==0:
                count+=1
        if count==2:
            print(i,end=" ")
x,y=map(int,input("enter x & y values:").split())
my(x,y)
```

```python
#count vowels
def my(string):
    count=0
    for i in string:
        if i in "AEIOUaeiou":
            count+=1
    print(count)
st=input("enter a string: ")
my(st)
```

```python
#frequency element in list
def my(l,x):
    m=len(l)
    count=0
    for i in range(m):
        if l[i]==x:
            count+=1
    print(f"frequency of element {x} is: ",count)
n=int(input("enter length of list: "))
li=[]
for i in range(n):
    z=int(input(f"enter the element{i} :"))
    li.append(z)
y=int(input("enter the freq element: "))
my(li,y)
```

```python
#armstrong number
x=int(input("enter a number: "))
def my(n):
    y=n
    res=0
    while n>0:
        temp=n%10
        res=res+(temp*temp*temp)
        n=n//10
    if res==y:
        print("armstrong number")
    else:
        print("non armstrong")
my(x)
```

```python
#strong number
import math
x=int(input("enter a number: "))
def my(n):
    y=n
    res=0
    while n>0:
        temp=n%10
        res=res+math.factorial(temp)
        n=n//10
    if res==y:
        print("strong number")
    else:
        print("not strong")
my(x)
```

```python
#remove duplicates in list
n=int(input("enter length of list:"))
li=[]
for i in range(n):
    x=int(input("enter the element:"))
    li.append(x)
def my(l):
    m=len(l)
    for i in range(m):
        for j in range(m-1):
            if l[j]>l[j+1]:
                l[j],l[j+1]=l[j+1],l[j]
    print(l)
    i=0
    while i<(len(l)-1):
        if l[i]==l[i+1]:
            del l[i]
        else:
            i+=1
    print(l)
my(li)
```

```python
#merge lists
n1=int(input("enter length of list:"))
l1=[]
for i in range(n1):
    x=int(input("enter the element:"))
    l1.append(x)
n2=int(input("enter length of list:"))
l2=[]
for i in range(n2):
    y=int(input("enter the element:"))
    l2.append(y)
def my(li1,li2):
    li1.extend(li2)
    print(li1)
my(l1,l2)
```

```python
#reverse lists
n1=int(input("enter length of list:"))
l1=[]
for i in range(n1):
    x=int(input("enter the element:"))
    l1.append(x)
def my(li):
    i=0
    j=len(li)-1
    while i<j:
        li[i],li[j]=li[j],li[i]
        i+=1
        j-=1
    print(li)
my(l1)
```
