```python
# postive or negative number
a=int(input("enter a:"))
if a>0:
    print("a is positive")
elif a<0:
    print("a is negative")
else:
    print("a is zero")
```

```python
#even or odd
a=int(input("enter a value:"))
if a%2==0:
    print("a is even")
else:
    print("a is odd")
```

```python
#sum of numbers in given range
a=int(input("enter the value of a:"))
b=int(input("enter the value of b:"))
sum=0
for i in range(a,b+1):
    sum+=i
print(sum)
```

```python
#prime number
a=int(input("enter the value of a:"))
count=0
for i in range(1,a+1):
    if a%i==0:
        count+=1
if count==2:
    print("prime number")
else:
    print("non prime number")
```

```python
#leap year
a=int(input("enter a year"))
if (a%4==0 and a%100!=0) or (a%400==0):
    print("leap year")
else:
    print("non leap year")
```

```python
#prime numbers in given range
a=int(input("enter the value of a:"))
b=int(input("enter the value of b:"))
for i in range(a,b+1):
    count=0
    for j in range(1,i+1):
        if i%j==0:
            count+=1
    if count==2:
        print(i,end=" ")
```

```python
#sum of digits
a=int(input("enter the value of a:"))
s=0
while a!=0:
    temp=int(a%10)
    s=s+temp
    a=a/10
print(s)
```

```python
#reserve a number
a=int(input("enter a :"))
n=0
while a>0:
    temp=a%10
    n=(n*10)+temp
    a=a//10
print(n)
```

```python
#palindrome number
a=int(input("enter a:"))
y = a
n=0
while a>0:
    temp=a%10
    n=(n*10)+temp
    a=a//10
if y==n:
    print("palindrome")
else:
    print("non palindrome")
```

```python
#armstrong number
a=int(input("enter a:"))
y=a
s=0
while a>0:
    temp=a%10
    s=s+(temp*temp*temp)
    a=a//10
if y==s:
    print("armstrong")
else:
    print("non armstrong")
```

```python
#factorial number
a=int(input("enter a:"))
m=1
for i in range(1,a+1):
    m=m*i
print(m)
```

```python
#strong number
a = int(input("enter a:"))
y = a
s = 0
while a > 0:
    m=1
    temp = a%10
    for i in range(1,temp+1):
        m = m*i
    s = s + m
    a = a // 10
if y == s:
    print("strong number")
else:
    print("non strong number")
```

```python
#power of a number
a=int(input("enter the value of exponential:"))
b=int(input("enter the value of power:"))
out=1
for i in range(b):
    out=out*a
print(out)
```

```python
#perfect number
a = int(input("enter a:"))
s=0
for i in range(1,a):
    if a%i==0:
        s=s+i
if a==s:
    print("perfect number")
else:
    print("non perfect number")
```

```python
#abundant number
a = int(input("enter a:"))
y=a
s=0
for i in range(1,a):
    if a%i==0:
        s=s+i
if s>y:
    print("abundant number")
else:
    print("non abundant number")
```
